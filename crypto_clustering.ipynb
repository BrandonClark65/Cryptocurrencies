# Clustering Crypto

```python
# Initial imports
import pandas as pd
from path import Path
import hvplot.pandas
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
crypto_df = pd.read_csv("Resources/crypto_data.csv")
crypto_df.head(10)
```

```python
# Keep all the cryptocurrencies that are being traded.
crypto_df = crypto_df[(crypto_df["IsTrading"]==True)]
crypto_df.head(10)
```

```python
# Remove the "IsTrading" column. 
crypto_df = crypto_df.drop(columns=["IsTrading"])
crypto_df.head(10)
```

```python
# Remove rows that have at least 1 null value.
crypto_df = crypto_df.dropna()
crypto_df.head(10)
```

```python
# Keep the rows where coins are mined.
crypto_df = crypto_df[crypto_df["TotalCoinsMined"] > 0]
crypto_df.head(10)
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
coin_names = crypto_df[["CoinName", "Unnamed: 0"]].copy()
coin_names_df = coin_names.set_index("Unnamed: 0")
coin_names_df.index.names = [None]
coin_names_df.head()
```

```python
# Set Unnamed: 0 as index
crypto_df = crypto_df.set_index("Unnamed: 0")
crypto_df.index.names = [None]
crypto_df.head()
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
crypto_df = crypto_df.drop(columns = ["CoinName"])
crypto_df.head(10)
```

```python
# Use get_dummies() to create variables for text features.
X = pd.get_dummies(crypto_df, columns = ["Algorithm", "ProofType"])
X.head(10)
```

```python
# Standardize the data with StandardScaler().
X_scaled = StandardScaler().fit_transform(X)
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Create a DataFrame with the three principal components.
pcs_df = pd.DataFrame(data=X_pca, index = X.index, columns=["PC1", "PC2", "PC3"])
pcs_df.head()
```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.
# Store values of k to plot
inertia = []
k = list(range(1,11))

# Looking for best k value
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
```

```python
# Define a DataFrame to plot the Elbow Curve using hvplot
elbow_data = {"k":k, "inertia":inertia}
df_elbow = pd.DataFrame(elbow_data)
# Plot the dataframe
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

Running K-Means with `k=4`

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)
```
